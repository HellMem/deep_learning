```
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
# To display in external window 
  
image = cv2.imread("dog.jpeg", 1) 
# Loading the image 
  

cubic = cv2.resize(image, (150, 150), interpolation= cv2.INTER_CUBIC)
inter_area = cv2.resize(image, (150, 150), interpolation= cv2.INTER_AREA)
  
  
Titles =["Original", "Inter Area", "Cubic"] 
images =[image, inter_area, cubic] 
count = 3
  
for i in range(count): 
    plt.subplot(2, 2, i + 1) 
    plt.title(Titles[i]) 
    plt.imshow(images[i]) 
  
plt.show() 
```
